```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Draw sample from a distribution  
loc, scale = 2, 0.1
samples = np.random.gumbel(loc, scale, 1000)    

# Perform the transformation on the samples here ...
count, bins, ignored = plt.hist(samples, 100, density = True)
plt.show()
```

```python
log_samples = np.log(samples)
count, bins, ignored = plt.hist(log_samples, 100, density = True)
plt.show()
```

```python
over_samples = 1/samples
count, bins, ignored = plt.hist(over_samples, 100, density = True)
plt.show()
```

```python
sqrt_samples = np.sqrt(samples)
count, bins, ignored = plt.hist(sqrt_samples, 100, density = True)
plt.show()
```

```python
e_samples = np.exp(samples)
count, bins, ignored = plt.hist(e_samples, 100, density = True)
plt.show()
```

```python
x_samples = samples**2
count, bins, ignored = plt.hist(x_samples, 100, density = True)
plt.show()
```
